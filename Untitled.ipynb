```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import optimize

def f_gauss(x, A, B,C,sigma):
    return A*np.exp(-(x-B)**2/(2*sigma**2))+C

def f_gauss2(x, A1, B1, C1,sigma1,A2,B2,C2,sigma2):
    return A1*np.exp(-(x-B1)**2/(2*sigma1**2))+A2*np.exp(-(x-B2)**2/(2*sigma2**2))

X1 = list(range(len(index)))
Y1 = y_tested.test_count
          
A11, B11, C11,sigma11,A12,B12,C12,sigma12= optimize.curve_fit(f_gauss2, X1, Y1)[0]
x1 = np.arange(0, 541,1)
y1 = A11*np.exp(-(x1-B11)**2/(2*sigma11**2))+A12*np.exp(-(x1-B12)**2/(2*sigma12**2))+C11+C12
plt.plot(x1, y1, "purple")




X = list(range(len(index)))

Y2 = y_confirmed.date_count
        
A21,B21,C21,sigma21,A22, B22,C22,sigma22= optimize.curve_fit(f_gauss2, X, Y2)[0]
x2 = np.arange(0, 541,1)
y2 = A21*np.exp(-(x2-B21)**2/(2*sigma21**2))+A22*np.exp(-(x2-B22)**2/(2*sigma22**2))+C21+C22

plt.plot(x1, y1, "purple")
plt.scatter(X,Y1)
plt.plot(x2, y2*180000/2500, "green")
plt.scatter(X,Y2*180000/2500)

plt.title("test")
plt.xlabel('x')
plt.ylabel('y')

plt.show()
```

```python
growth = [0]
confirmed = y_confirmed.date_count
for i in range(len(y_confirmed.date_count)-1):
    rate = (confirmed[i + 1] - confirmed[i])
    growth.append(rate)
```

```python

plt.figure(figsize=(8,5))
plt.plot(index[0:541], growth[0:541],color='green',linestyle = 'solid')
plt.xticks(index[0:len(index):30], 
           rotation = 90)
plt.title("Growth rate")
plt.xlabel("Date")
plt.ylabel("Growth")
plt.show()
```

```python
# 考虑到delta病毒带来的感染率显著上升 我们认为以感染人数或者阳性率来判断一项措施是否有效是不合适的
# 因为一项措施施行时可能晚于delta病毒出现，这将导致 即便该措施可能是有效的 但确诊人数或者阳性率依然很高
# 我们更关注一项措施施行后，一段时间里（比如一个月）每日新增确诊人数的增速(growth rate)。我们将它称为growth rate
# 比如day1开始施行边境封锁， day1 当天确诊100人， day30 当天确诊10人，这一个月的growth rate为 (10-100)/30 = - 3 人/天
# 也就是这一个月内，每过一天，新增确诊的人数平均会少 3 个人
# 如果一段时间里每天的新增确诊人数保持不变 那么增速growth rate为 0
# 该数值越小，表明该周期内（一个周期可以是一个月、两个月等等）防疫效果越好
# 我们将通过比较两个周期各自的growth rate的值，判断后一个周期跟前一个周期比，每日新增确诊增速是否放缓，
# 也就是防疫效果是否更好了，然后量化、可视化这个防疫效果的变化
# 我们希望一个好的措施能让该措施施行后一段时间内的growth rate 明显低于该措施施行前的growth rate

# 由于戴口罩、隔离边境等措施一瞬间就可以完成 所以以该政策实施前后同一长度时间段（一个月或两个月）的growth rate作比较
# 由于增加检测量、打疫苗是连续的过程 所以将它们实施的过程分为前期与后期 比较前后期的growth rate

print("戴口罩前二月growth rate: "+str(sum(growth[228:288])/len(growth[228:288])))#口罩前一月
print("戴口罩后二月growth rate: "+str(sum(growth[288:288+60])/len(growth[288:288+60])))#口罩后一月

print("增加测试前期growth rate: ")
print(sum(growth[420:480])/len(growth[420:480]))#增加测试初期 或打疫苗初期
print("增加测试后期growth rate: ")
print(sum(growth[480:541])/len(growth[480:541]))#增加测试后期 或打疫苗后期

# 由于测试前期与测试后期与疫苗前期、疫苗后期基本重合 我们将其视为一种组合
# 由于某个封锁边境措施也发生在这个阶段 我们将根据其他封锁边境措施的效果判断是否将此次封锁边境也纳入组合
# 虽然戴口罩、封锁边境的措施可以让接下来的一段时间里growth rate降低，但它不会一直降
# 新增确诊会收敛到一个相对低的值，也就是growth rate会在政策实行一段时间后趋于0，每日新增不再明显变化
# 我们正好利用这一点 来比较不同措施带来的防疫效果
# 在疫苗以及增加测试的时间范围内，由于戴口罩与封锁边境也一直持续，但距离口罩、封锁政策施行已经过了相当久的时间
# 我们倾向于认为在增加检测、打疫苗两政策施行时间范围内，戴口罩、封锁边境的growth rate已经趋近于0
# 该时间段内growth rate的变化经历了由delta病毒导致的上升，以及后期增加测试与打疫苗效果逐渐显现导致的下降。

# 我们将每个措施前后growth rate的变化值做成barchart，比较每个措施或组合的防疫效果。
measure_list = ("mask","test and vaccine")
growth_rate_diff = [sum(growth[288:288+60])/len(growth[288:288+60])-sum(growth[228:288])/len(growth[228:288]),sum(growth[480:541])/len(growth[480:541])-sum(growth[420:480])/len(growth[420:480])]


plt.figure(figsize=(8,5))
plt.barh(measure_list, growth_rate_diff,color='green')
plt.ylim()
plt.title("Growth rate")
plt.xlabel("Date")
plt.ylabel("Growth")
plt.show()

# 如下图 这是说 在test与vaccine后期平均每天比前期少新增13个人
```
